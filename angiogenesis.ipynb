```python
import numpy as np
import matplotlib.pyplot as plt

# Ler arquivos de dados
arquivos_dados = ['v300-p15-ve15', 'v300-p15-ve5', 'v300-p5-ve15', 'v300-p5-ve5']

# Informações personalizadas para cada legenda
informacoes_legenda = ['v=26 \u03B1P(T)=1,5 Ti=15', 'v=26 \u03B1P(T)=1,5 Ti=5', 'v=26 \u03B1P(T)=0,5 Ti=15', 'v=26 \u03B1P(T)=0,5 Ti=5 ']

# Criar a figura
fig, ax = plt.subplots(dpi=500)

# Cores e símbolos para cada conjunto de dados
cores = ['red', 'blue', 'green', 'orange']
simbolos = ['o', 's', 'v', '^']

# Fator de escala
fator_escala = 26/60000


# Iterar sobre os arquivos de dados
for i, arquivo in enumerate(arquivos_dados):
    # Ler dados
    dados = np.loadtxt(arquivo, skiprows=1, delimiter=';')
    
    # Declarando variáveis
    x = dados[:, 0] * fator_escala  # Multiplicar o eixo x pelo fator de escala
    y = dados[:, 2] * 1.25
    
    # Plotar gráfico de dispersão para cada conjunto de dados
    ax.scatter(x, y, color=cores[i], marker=simbolos[i], label=informacoes_legenda[i])
    
# Definir rótulos dos eixos
ax.set_xlabel('Tempo (hr)')
ax.set_ylabel('Diâmetro dos vasos (\u03BCm)')

# Título do gráfico
ax.set_title('Diâmetro dos vasos vs Tempo')

# Legenda
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')


#Salvando imagem
plt.savefig('diameter-time-300.png', format='png')

# Exibir o gráfico
#plt.show()
```
