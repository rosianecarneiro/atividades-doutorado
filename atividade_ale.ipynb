```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
k = 1
j = 1
#m = 1

Ts = np.arange(0.01, 2.1, 0.01)
res = []
Mres =[]
for T in Ts:
    B=1/(k*T)
    res.append(B)
    M = np.size(res)
    Mres.append(M)
    #print(M)

plt.figure()
plt.errorbar(Mres,res,uplims=True,ls="-",lw=2,marker="o",ms=7.5,color="k")
plt.xlabel(r"$\rho$",fontsize=20)
plt.show()
```
